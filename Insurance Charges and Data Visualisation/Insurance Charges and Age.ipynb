```python
#import CSV file to jupyter

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import the data
df = pd.read_csv("insurance.csv")

# Plotting a scatterplot with age on x-axis and charges on y-axis. 
x = df['age'].values
y = df['charges'].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Create a model
linreg_model = LinearRegression()

# Make the model compatible
X = df['age'].values.reshape(-1,1)
y = df['charges'].values

# Fit the model 
linreg_model.fit(X,y)
```

```python
# Plot another scatter plot with the best-fit line.

y_pred = linreg_model.predict(X)
plt.figure()
plt.scatter(x,y,color = 'b')
plt.plot(x, y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Predict an unknown value with the same scatter plot
unk_x = [[22]] 

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = linreg_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print("Predicted insurance charge for 22 years old is ",linreg_model.predict(unk_x))
```
